```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/combined/recession_probability.csv', parse_dates=['date'])
```

```python
display(df)
display(df.info())
display(df.describe())
```

```python
missing = df.isnull().sum()
display(missing[missing > 0])
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['date'], df['recession_probability'], label='Recession Probability')
plt.xlabel('Date')
plt.ylabel('Probability (%)')
plt.title('Recession Probability Over Time')
plt.legend()
plt.show()

```

```python
econ_cols = [
    'recession_probability',
    'CPI',
    'INDPRO',
    'unemployment_rate',
    'PPI',
    'OECD_CLI_index',
    'share_price',
    'CSI_index',
    '1_year_rate',
    '3_months_rate',
    '6_months_rate',
    '10_year_rate'
]
plt.figure(figsize=(10,7))
sns.heatmap(df[econ_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Economic Indicators')
plt.show()
```

```python
for col in econ_cols[1:]:
    plt.figure(figsize=(10,4))
    plt.plot(df['date'], df[col])
    plt.xlabel('Date')
    plt.ylabel(col)
    plt.title(f'{col} Over Time')
    plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='unemployment_rate', y='recession_probability', data=df)
plt.title('Unemployment Rate vs. Recession Probability')
plt.show()
```

```python
plt.plot(df['date'], df['CPI'], label='CPI')
plt.plot(df['date'], df['INDPRO'], label='Industrial Production')
plt.plot(df['date'], df['PPI'], label='PPI')
plt.plot(df['date'], df['OECD_CLI_index'], label='OECD CLI Index')
plt.plot(df['date'], df['share_price'], label='Share Price')
plt.xlabel('Date')
plt.legend()
plt.show()
```

```python
for indicator in econ_cols[1:]:
    fig, ax1 = plt.subplots(figsize=(10,5))
    ax2 = ax1.twinx()

    ax1.plot(df['date'], df['recession_probability'], color='tab:blue', label='Recession Probability')
    ax2.plot(df['date'], df[indicator], color='tab:orange', label=indicator)

    ax1.set_xlabel('Date')
    ax1.set_ylabel('Recession Probability', color='tab:blue')
    ax2.set_ylabel(indicator, color='tab:orange')
    plt.title(f'Recession Probability and {indicator} Over Time')

    fig.tight_layout()
    plt.show()
```
